```python
import cv2
import numpy as np
import imgaug as ia
import imgaug.augmenters as iaa
import matplotlib.pyplot as plt
import random
import os
from PIL import Image

# 이미지와 라벨 입력 시 그려줌
def img_res(im,txt):
    bbox=[] #라벨 리스트

    #이미지 읽어오기
    img = cv2.imread(im)

    #라벨 읽어오기
    gt = np.loadtxt(txt, dtype = float, delimiter = ' ')

    #라벨이 없다면 empty값 출력
    if len(gt) == 0:
        return img,0
    if gt.ndim==1:
        gt=np.array([gt])
    
    #이미지 사이즈 저장
    img_x, img_y = img.shape[1],img.shape[0]
    
    input_img = img[np.newaxis, :, :, :]

    #라벨 리스트에 라벨 추가
    for i in range(gt.shape[0]):
        bbox.append(ia.BoundingBox(x1 = gt[i][1]*img_x-(gt[i][3]*img_x)/2, y1 = gt[i][2]*img_y-(gt[i][4]*img_y)/2, x2 = gt[i][1]*img_x+(gt[i][3]*img_x)/2, y2 = gt[i][2]*img_y+(gt[i][4]*img_y)/2))

    #이미지와 라벨 변환
    output_img, output_bbox = iaa.Affine(scale = 1)(images = input_img, bounding_boxes = bbox)

    #초기에 한 번 실행
    draw = output_bbox[0].draw_on_image(output_img[0], size = 2, color = [0,0,255])

    #추가적인 라벨이 있다면 마저 그려줌
    for k in range(1,gt.shape[0]):
        draw = output_bbox[k].draw_on_image(draw, size = 2, color = [0,0,255])

    
    return draw,gt.shape[0]



root_dir = "C:/scr/darknet/test_dataset_yhg" # 디렉토리

possible_img_extension = ['.jpg', '.jpeg', '.JPG', '.bmp', '.png'] # 이미지 확장자들

#전체파일 이름
name = "32_avi_128_yhg"

#해당 파일 생성

# os.mkdir("C:/scr/darknet/data/"+name)
# os.mkdir("C:/scr/darknet/data/"+name +"/"+name+"_0")
# os.mkdir("C:/scr/darknet/data/"+name + "/"+name+"_1")
# os.mkdir("C:/scr/darknet/data/"+name + "/"+name+"_2")
# os.mkdir("C:/scr/darknet/data/"+name + "/"+name+"_3")



# 경로의 파일 하나씩 읽어오기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
        for file_name in files:
            if os.path.splitext(file_name)[1] in possible_img_extension:
                
                img_path = root + '/' + file_name
                img_path = img_path.replace('\\', '/') # \는 \\로 나타내야함
                img = cv2.imread(img_path)
                draw,num  = img_res(img_path,img_path[:-4]+'.txt')
                cv2.imwrite("C:/scr/darknet/data/" +name +"/"+file_name,draw)
                
                
    
```
